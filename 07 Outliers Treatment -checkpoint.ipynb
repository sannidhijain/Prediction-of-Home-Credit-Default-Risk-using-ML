```python
ds.shape
```

```python
q1=ds.quantile(0.25)
q3=ds.quantile(0.75)
iqr=q3-q1
ll=q1-iqr*10
ul=q3+iqr*10
```

```python
do=ds[~((ds<ll)|(ds>ul)).any(axis=1)]
```
