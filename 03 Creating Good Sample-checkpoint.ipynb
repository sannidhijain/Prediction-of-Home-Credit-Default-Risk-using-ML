```python
import pandas as pd
import numpy as np
pd.options.display.max_columns=None
pd.options.display.max_rows=None

import scipy.stats as stats
```

```python
df_pop = pd.read_csv(r'application_train.csv\application_train.csv')
df_pop.head()
```

```python
def num_sample_pop_test(col,df_pop,df_samp,alpha = 0.05,verbose=False):
    pop = df_pop[col]
    samp = df_samp[col]
    
    pop_mean = pop.mean()
    samp_mean = samp.mean()
    
    pop_std = pop.std()
    samp_std = samp.std()
    
    x_bar = samp_mean
    n = len(samp)
    sigma = pop_std

    z_alpha_by_2 = stats.norm.isf(alpha/2)

    upper = x_bar + z_alpha_by_2 * sigma / n**0.5
    lower = x_bar - z_alpha_by_2 * sigma / n**0.5
    #print('95% Conf interval:', lower, upper)

    if lower<=pop_mean<=upper:
        if verbose: print(f'{col} Sample reperesents Population')
        return (col,1)
    else:
        if verbose: print(f'{col} Sample and Population are different')
        return (col,0)
```

```python
def get_num_not_good(df_pop,df_samp,alpha=0.05,verbose=False):
    num_cols = df_pop.select_dtypes(np.number)
    l = []
    for col in num_cols:
        l.append(num_sample_pop_test(col,df_pop,df_samp,alpha=alpha,verbose=verbose))
    num_not_good = [i for i in l if i[1]==0]
    return num_not_good
```

```python
def cat_samp_pop_test(col,df_pop,df_samp,alpha=0.05,verbose=False):
    observed = df_samp[col].value_counts()
    expected = df_pop[col].value_counts(normalize=True)*np.sum(observed)
    
    cats_not_in_samp = []
    for cat in expected.index:
        if cat not in observed.index:
            cats_not_in_samp.append(cat)
    observed = observed.append(pd.Series([0]*len(cats_not_in_samp),index=cats_not_in_samp,dtype=np.number))
    
    #display(pd.DataFrame([observed,expected],index=['observed','expected']))
    
    chi2_stat,p_val = stats.chisquare(f_obs = observed, f_exp = expected)
    
    if p_val > alpha/2:
        if verbose: print(f'{col} Sample represents the Population')
        return (col,1)
    else:
        if verbose: print(f'{col} Sample and Population are different')
        return (col,0)
```

```python
def get_cat_not_good(df_pop,df_samp,alpha=0.05,verbose=False):
    cat_cols = df_pop.select_dtypes(object)
    l = []
    for col in cat_cols:
        l.append(cat_samp_pop_test(col,df_pop,df_samp,alpha=alpha,verbose=verbose))
    cat_not_good = [i for i in l if i[1]==0]
    return cat_not_good
```

```python
def get_num_cat_not_good(df_pop,df_samp,alpha=0.05,verbose=False):
    d = df_pop.select_dtypes(np.number).apply(lambda col:col.nunique())
    d = d[d<5]
    
    num_cat_cols = d.index
    #categorical test
    l = []
    for col in num_cat_cols:
        l.append(cat_samp_pop_test(col,df_pop,df_samp,alpha=alpha,verbose=verbose))
    num_cat_not_good = [i for i in l if i[1]==0]
    return num_cat_not_good
```

```python
def generate_good_sample(df_pop,sample_size=30000):
    count = 0
    limit = df_pop.shape[0]**0.5
    while count<limit:
        df_samp = df_pop.sample(sample_size)
        pop = df_pop.drop('SK_ID_CURR',axis=1)
        samp = df_samp.drop('SK_ID_CURR',axis=1)
        num = get_num_not_good(pop,samp)
        cat = get_cat_not_good(pop,samp)
        num_cat = get_num_cat_not_good(pop,samp)
        print('Num:',num)
        print('Cat:',cat)
        print('Num_Cat:',num_cat)
        
        if len(num)+len(cat)+len(num_cat) == 0:
            return df_samp
    else:
        print('Sample generation not possible')
```

```python
df_samp1 = generate_good_sample(df_pop)
```

```python
df_samp1.to_csv('sample_train_good_30k_with_user_ids.csv')
```
