##### Functions for univariante analysis for Categorical and Categorical vs Target

```python
def univariante_cat_vs_target(df,col,col_labels={}):
    table = pd.crosstab(df[col],df['TARGET'])
    table.columns = table.columns.map({0:'Paid',1:'Not Paid'})
    table.plot(kind='bar')
    if col_labels:
        plt.xticks(ticks = table.index,labels = table.index.map(col_labels))
    plt.ylabel('Count')
    plt.title(f'{col} vs TARGET')
    plt.show()
```

```python
def display_value_counts_crosstabs_cat_vs_target(df,col,col_labels={}):
    print(f'Value Counts for {col}:')
    display(df[col].value_counts())
    print('-'*100)
    print(f'Percentage of each category for {col}:')
    t = (df[col].value_counts(normalize=True)*100)
    display(t)
    print('-'*100)
    
    if col_labels:
        plt.pie(t, labels=t.index.map(col_labels),autopct='%.2f',shadow=True,radius=0.9)
    else:
        plt.pie(t,labels= t.index,autopct='%.2f',shadow=True,radius=0.9)
    plt.title(col)
    plt.show()

    print('Cross tab:')
    display(pd.crosstab(df[col],df['TARGET']))
    print('-'*100)

    print('Cross tab normalized:')
    display(pd.crosstab(df[col],df['TARGET'],normalize=0)*100)
         
    univariante_cat_vs_target(df,col,col_labels)
```
