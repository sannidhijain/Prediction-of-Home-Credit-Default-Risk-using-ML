```python
import pandas as pd
import numpy as np
pd.options.display.max_columns=None
pd.options.display.max_rows=150
pd.options.display.max_colwidth = None

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15,8)

import seaborn as sns
```

```python
df = pd.read_csv('sample_train_good_40k.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
col_desc = pd.read_csv('HomeCredit_columns_description.csv')
description = col_desc[col_desc['Table']=='application_{train|test}.csv'][['Row','Description','Special']]
description
```

```python
def get_description_of_column(col,df=description):
    return description[description['Row']==col]
```

```python
dtypes = df.apply(lambda col: col.dtype)
null_count = df.isnull().sum()
nuniques = df.apply(lambda col:col.nunique())

df_info = pd.DataFrame({'dtypes':dtypes,'null_count':null_count,'nuniques':nuniques})
df_info
```

```python
num_cat_col = df_info[(df_info['dtypes']!='object')&(df_info['nuniques']<4)].index
num_col = df_info[(df_info['dtypes']!='object')&(df_info['nuniques']>=4)].index
cat_col = df_info[df_info['dtypes']=='object'].index

print(len(num_col)+len(cat_col)+len(num_cat_col))

cat_col = list(num_cat_col) + list(cat_col)
```

```python
print('Number of Categorical columns:',len(cat_col))
print()
print(cat_col)
```

```python
def cat_vs_cat_plot(df,col1,col2):
    t = pd.crosstab(df[col1],df[col2])
    print(f'Crosstab {col1} vs {col2}:')
    display(t)
    t2 = pd.crosstab(df[col1],df[col2],normalize=0)
    print(f'Crosstab Normalized {col1} vs {col2}:')
    display(t2)
    t.plot(kind='bar')
    plt.title(f'{col1} vs {col2}')
    plt.ylabel('Count')
    plt.show()
```

#### Communication Related

```python
comm_cols = ['FLAG_MOBIL', 'FLAG_EMP_PHONE', 'FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE', 'FLAG_PHONE', 'FLAG_EMAIL']
```

```python
for i in range(len(comm_cols)):
    col1 = comm_cols[i]
    for j in range(i+1,len(comm_cols)):
        col2 = comm_cols[j]
        print(f'{col1} Description:',get_description_of_column(col1)['Description'].values[0])
        print(f'{col2} Description:',get_description_of_column(col2)['Description'].values[0])
        cat_vs_cat_plot(df,col1,col2)
        print('-'*100)
```
